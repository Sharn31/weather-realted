# Visiualizations

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create 3-row subplot layout
fig = make_subplots(
    rows=3, cols=1,
    subplot_titles=[
        "Temperature Distribution by Disease",
        "Humidity Distribution by Disease",
        "Wind Speed Distribution by Disease"
    ]
)

# 1.Temperature
fig.add_trace(
    go.Box(
        x=df['prognosis'],
        y=df['Temperature (C)'],
        name="Temperature",
        marker_color='tomato'
    ),
    row=1, col=1
)

# 2.Humidity
fig.add_trace(
    go.Box(
        x=df['prognosis'],
        y=df['Humidity'],
        name="Humidity",
        marker_color='royalblue'
    ),
    row=2, col=1
)

#3. Wind Speed
fig.add_trace(
    go.Box(
        x=df['prognosis'],
        y=df['Wind Speed (km/h)'],
        name="Wind Speed",
        marker_color='mediumseagreen'
    ),
    row=3, col=1
)

# Layout customization
fig.update_layout(
    height=1200,
    width=1000,
    showlegend=False,
    title=dict(
        text="Climate Factors Distribution by Disease",
        x=0.5,  # center title
        xanchor='center'
    )
)

# Rotate x-axis labels
fig.update_xaxes(tickangle=45)

fig.show()
fig.write_html("climate_factors_distribution.html")
```

```python
bins   = list(range(0, 101, 10))
labels = [f"{bins[i]}–{bins[i+1]-1}" for i in range(len(bins)-1)]
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
df['age_group'] = df['age_group'].cat.set_categories(labels, ordered=True)
counts = (
    df.groupby(['age_group','prognosis']).size().reset_index(name='count')
)

# 3) Compute total cases *in each age_group*
totals = (
    df
    .groupby('age_group')
    .size()
    .reset_index(name='total')
)

# 4) Merge them & compute rate “per 100 people in that bin”
counts = counts.merge(totals, on='age_group')
counts['per_100'] = counts['count'] / counts['total'] * 100
```

```python
import plotly.express as px

fig = px.bar(
    counts,
    x='age_group',
    y='per_100',
    color='prognosis',
    barmode='group',
    title='Disease Rate by 10-Year Age Group (per 100)',
    text='per_100'
)

# ✅ Move width & height into update_layout
fig.update_layout(
    title=dict(text='Disease Rate by 10-Year Age Group (per 100)', x=0.5),
    xaxis_title='Age Group',
    yaxis_title='Cases per 100 People',
    legend_title_text='Disease',
    xaxis={'categoryorder': 'array', 'categoryarray': labels},
    width=1000,     # plot width
    height=600      # plot height
)

# ✅ Keep text & bar style updates here
fig.update_traces(
    texttemplate='%{text:.2f}',
    textposition='outside'
)

fig.show()
fig.write_html("Disease Rate by 10-Year Age Group.html")

```

```python
# Bin weather variables for categorical analysis
df['temp_bin'] = pd.cut(df['Temperature (C)'], bins=[0, 20, 25, 30, 35, 100], labels=['<20°C', '20-25°C', '25-30°C', '30-35°C', '>35°C'])
df['humidity_bin'] = pd.cut(df['Humidity'], bins=[0, 30, 50, 70, 100], labels=['Low', 'Medium', 'High', 'Very High'])
df['wind_bin'] = pd.cut(df['Wind Speed (km/h)'], bins=[0, 10, 20, 30, 100], labels=['Low', 'Medium', 'High', 'Very High'])

# Create heatmaps for each combination
plt.figure(figsize=(14, 10))
temp_disease = pd.crosstab(df['temp_bin'], df['prognosis'], normalize='index') * 100
sns.heatmap(temp_disease, annot=True, fmt='.1f', cmap='YlOrRd')
plt.title('Disease Prevalence (%) by Temperature Range')
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 10))
humidity_disease = pd.crosstab(df['humidity_bin'], df['prognosis'], normalize='index') * 100
sns.heatmap(humidity_disease, annot=True, fmt='.1f', cmap='YlGnBu')
plt.title('Disease Prevalence (%) by Humidity Level')
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 10))
wind_disease = pd.crosstab(df['wind_bin'], df['prognosis'], normalize='index') * 100
sns.heatmap(wind_disease, annot=True, fmt='.1f', cmap='Greens')
plt.title('Disease Prevalence (%) by Wind Speed')
plt.tight_layout()
plt.show()
fig.write_html("Disease Prevalence by Temperature Range.html")
```

```python
df['Gender_cat'] = df['Gender'].map({0: 'Male', 1: 'Female'})

#  make it a true ordered/category dtype
df['Gender_cat'] = df['Gender_cat'].astype('category').cat.set_categories(['Male','Female'])

# 1) Bin ages as before
bins   = list(range(0, 101, 10))
labels = [f"{bins[i]}–{bins[i+1]-1}" for i in range(len(bins)-1)]
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
df['age_group'] = df['age_group'].cat.set_categories(labels, ordered=True)

# 2) Compute rates per 100
counts = (
    df
    .groupby(['Gender_cat','age_group','prognosis'],observed=True)
    .size()
    .reset_index(name='count')
)
totals = (
    df
    .groupby(['Gender_cat','age_group'],observed=True)
     .size()
    .reset_index(name='total')
)
rates = counts.merge(totals, on=['Gender_cat','age_group'])
rates['per_100'] = rates['count'] / rates['total'] * 100
```

```python
import plotly.express as px

fig = px.bar(
    rates,
    x='age_group',
    y='per_100',
    color='prognosis',
    facet_col='Gender_cat',             # separate Male / Female panels
    barmode='group',
    text='per_100',
    title='Disease Rates by Age Group and Gender (per 100)',
    color_discrete_sequence=px.colors.qualitative.Set2,
    width=1000,                         # ✅ chart width
    height=400                          # ✅ chart height
)

# Layout customization
fig.update_layout(
    title=dict(text='Disease Rates by Age Group and Gender (per 100)', x=0.5),
    xaxis_title='Age Group',
    yaxis_title='Cases per 100 People',
    legend_title_text='Disease',
    bargap=0.15,
    margin=dict(t=100, b=80, l=80, r=80)
)

# Text and axis formatting
fig.update_traces(
    texttemplate='%{text:.1f}',
    textposition='outside'
)

fig.update_xaxes(categoryorder='array', categoryarray=rates['age_group'].unique())
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))  # Clean facet labels

fig.show()
fig.write_html("Disease Rates by Age Group and Gender.html")
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns

# Select diseases and assign colors dynamically
selected_diseases = ['Common Cold', 'Influenza', 'Heat Stroke']
palette = sns.color_palette("Set2", len(selected_diseases))

fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Plot each disease group
for i, disease in enumerate(selected_diseases):
    disease_data = df[df['prognosis'] == disease]
    ax.scatter(
        disease_data['Temperature (C)'], 
        disease_data['Humidity'], 
        disease_data['Wind Speed (km/h)'],
        c=[palette[i]],
        label=disease,
        alpha=0.75,
        s=60,                     # point size
        edgecolors='k'            # black edge for contrast
    )

# Axis labels and title
ax.set_xlabel('Temperature (°C)', labelpad=10)
ax.set_ylabel('Humidity (%)', labelpad=10)
ax.set_zlabel('Wind Speed (km/h)', labelpad=10)
ax.set_title('3D Scatter Plot: Weather Variables vs Disease Type', fontsize=14, pad=20)

# Legend and style tweaks
ax.legend(title='Disease', loc='best')
ax.view_init(elev=25, azim=135)  # better viewing angle
plt.tight_layout()
plt.show()

fig.write_html("Disease Rates by Age Group and Gender.html")
```

```python
import plotly.graph_objects as go

selected_diseases = ['Common Cold', 'Influenza', 'Heat Stroke']
colors = ['#66c2a5', '#fc8d62', '#8da0cb']  # similar to Seaborn Set2 palette

fig = go.Figure()

for i, disease in enumerate(selected_diseases):
    disease_data = df[df['prognosis'] == disease]
    fig.add_trace(
        go.Scatter3d(
            x=disease_data['Temperature (C)'],
            y=disease_data['Humidity'],
            z=disease_data['Wind Speed (km/h)'],
            mode='markers',
            marker=dict(
                size=6,
                color=colors[i],
                line=dict(width=1, color='black')
            ),
            name=disease
        )
    )

fig.update_layout(
    scene=dict(
        xaxis_title='Temperature (°C)',
        yaxis_title='Humidity (%)',
        zaxis_title='Wind Speed (km/h)'
    ),
    title=dict(
        text='3D Scatter Plot: Weather Variables vs Disease Type',
        x=0.5,            # center the title
        xanchor='center'
    ),
    width=900,
    height=700
)

# Save as interactive HTML
fig.write_html("Disease_3D_Scatter.html")
print("HTML file saved successfully!")

```
